```python
import pandas as pd 
import datetime 
from datetime import date, timedelta
import yfinance as yf 
import plotly.graph_objects as go


today = date.today()
end_date = today.strftime("%Y-%m-%d")
start_date = (date.today()- timedelta(days=730)).strftime("%Y-%m-%d")
```

```python
df = yf.download(
    'BTC-INR',
    start= start_date,
    end= end_date,
    progress=False)

df["Date"]= df.index
df = df[["Date", 'Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume']]
df.reset_index(drop=True, inplace=True)
print(df.head())
```

```python
df.shape
```

```python
import plotly.graph_objects as go
figure = go.Figure(data=[go.Candlestick(x=df['Date'],
                                        open=df['Open'], 
                                        high=df['High'],
                                        low=df['Low'], 
                                        close=df['Close'])])

figure.update_layout(title = 'Bitcoin Price Analysis', 
                     xaxis_rangeslider_visible=False)
figure.show()
```

```python
correlation = df.corr()
print(correlation["Close"].sort_values(ascending=False))
```

```python
from autots import AutoTS
#pd.options.display.float_format = '{:.2f}'.format  # sets the display format for floats
model = AutoTS(forecast_length=30, frequency='infer', ensemble='simple', verbose=0)
model = model.fit(df, date_col='Date', value_col='Close', id_col=None)

```

```python
pd.options.display.float_format = '{:.2f}'.format  # sets the display format for floats
prediction = model.predict()
forecast = prediction.forecast
print(forecast)
```
